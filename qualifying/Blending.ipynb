```python
import pandas as pd
import numpy as np
import glob
```

```python
path_train = './data/train'
path_test = './data/test'
path_saved_models = './data/saved_models'
path_saved_predictions = './data/saved_predictions'
```

```python
file_names = glob.glob(path_saved_predictions + "/*.csv")
combined_df = pd.DataFrame()

for i, file in enumerate(file_names):
    df = pd.read_csv(file)
    if i == 0:
        combined_df['id'] = df['id']
    combined_df = pd.concat([combined_df, df[['target']].rename(columns={"target":f"target_{i}"})], axis=1)

combined_df
```

```python
combined_df["target"] = combined_df.drop(columns=['id']).mean(axis=1)
combined_df
```

```python
combined_df[['id', 'target']].to_csv("i_love_democracy.csv", index=False)
```
